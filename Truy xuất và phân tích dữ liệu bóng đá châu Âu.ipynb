14. Liệt kê số trận thắng, thua và hòa của mỗi đội.

- Nếu một đội có số lượng bàn thắng trên sân nhà > số lượng bàn thắng trên sân khách của chính đội đó, thì đội đó sẽ "thắng" trong trận đấu này
- Nếu một đội có số lượng bàn thắng trên sân nhà < số lượng bàn thắng trên sân khách của chính đội đó, thì đội đó sẽ "thua" trong trận đấu này
- Nếu một đội có số lượng bàn thắng trên sân nhà = số lượng bàn thắng trên sân khách của chính đội đó, thì đội đó sẽ "hòa" trong trận đấu này
- Sử dụng câu lệnh "Case When"

```python
# conn.execute("""CREATE VIEW table_team_WLT
# AS
result_goal = pd.read_sql("""SELECT B.team_long_name, B.goal, COUNT(*) AS match_tally
FROM ( SELECT TH.team_long_name
,(CASE
WHEN MH.home_team_goal > MH.away_team_goal then 'Win'
WHEN MH.home_team_goal < MH.away_team_goal then 'Lose'
ELSE 'Tie'
END) AS goal
FROM Team AS TH
INNER JOIN Match AS MH
ON MH.home_team_api_id = TH.team_api_id

UNION ALL

SELECT TA.team_long_name
,(CASE
WHEN MA.away_team_goal > MA.home_team_goal then 'Win'
WHEN MA.away_team_goal < MA.home_team_goal then 'Lose'
ELSE 'Tie'
END) AS goal
FROM Team AS TA
INNER JOIN Match AS MA
ON MA.away_team_api_id = TA.team_api_id) AS B
GROUP BY B.team_long_name, B.goal;""",conn)
result_goal.head()
```

17. Liệt kê tỉ lệ phần trăm về số bàn ghi được của từng giải so với tổng số bàn của tất cả các trận

tạo bảng phụ từ view để tìm số bàn được từ home team và away team của mỗi giải

```python
conn.execute("""CREATE VIEW table_league
AS
SELECT L.name, M.season
,(M.home_team_goal + M.away_team_goal) AS goal
FROM League AS L
INNER JOIN Match AS M
ON M.league_id = L.id;""")
```

từ bảng phụ đó
em sub-query tìm tổng số các bàn thắng đã ghi trong tất cả trận đấu ở bàng phụ table_league

```python
result_percent_goal_league = pd.read_sql("""SELECT name
,SUM(goal)*100 / (SELECT SUM(goal) FROM table_league) AS 'Percent(%)'
FROM table_league
GROUP BY name;""",conn)

result_percent_goal_league
```

21. Liệt kê cầu thủ có số trận đấu nhiều nhất

```python
conn.execute("""CREATE VIEW table_player_match
AS
SELECT B.player_name, SUM(match_tally) AS match_total
FROM
( SELECT PH.player_name, COUNT(*) AS match_tally
FROM Player AS PH, Match AS MH
WHERE PH.player_api_id IN (MH.home_player_1, MH.home_player_2, MH.home_player_3
,MH.home_player_4, MH.home_player_5, MH.home_player_6
,MH.home_player_7, MH.home_player_8, MH.home_player_9
,MH.home_player_10, MH.home_player_11)
GROUP BY PH.player_name

UNION ALL

SELECT PA.player_name, COUNT(*) AS match_tally
FROM Player AS PA, Match AS MA
WHERE PA.player_api_id IN (MA.home_player_1, MA.home_player_2, MA.home_player_3
,MA.home_player_4, MA.home_player_5, MA.home_player_6
,MA.home_player_7, MA.home_player_8, MA.home_player_9
,MA.home_player_10, MA.home_player_11)
GROUP BY PA.player_name) AS B
GROUP BY B.player_name;
""")
```

```python
result_match_player = pd.read_sql("""SELECT *
FROM table_player_match
WHERE match_total =
(SELECT max(match_total)
FROM table_player_match);""",conn)

result_match_player
```

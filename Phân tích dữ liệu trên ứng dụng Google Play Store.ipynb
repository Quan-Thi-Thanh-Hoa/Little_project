# **Làm việc với dữ liệu DateTime trong Pandas để phân tích dữ liệu trên ứng dụng Google Play Store**

```python
import pandas as pd
df_google_playstore = pd.read_csv(r'C:\Users\Quan Thi Thanh Hoa\AppData\asm2.ipynb\DSP302\Google_Playstore.csv')
df_google_playstore.head()
```

```python
# Yêu cầu: Chuyển đổi dữ liệu của 3 cột Released, Last Updated và Scraped Time về dạng datetime
datetime_cols = ['Released', 'Last Updated', 'Scraped Time']
for col in datetime_cols:
    df_google_playstore[col] = pd.to_datetime(df_google_playstore[col])
```

```python
df_google_playstore[datetime_cols].fillna(0, inplace=True)
df_google_playstore[datetime_cols].isna().sum()
```

```python
datetime_cols = ['Released', 'Last Updated', 'Scraped Time']
df_google_playstore[datetime_cols]
```

```python
# Tìm hiểu tháng nào thường phát hành ứng dụng nhiều nhất
df_google_playstore['month'] = df_google_playstore['Released'].apply (lambda x: str(x).split('-')).str[1]
df_google_playstore.groupby(['month']).count()['Released'].sort_values(ascending =False)
```

```python
# Yêu cầu: Tìm hiểu năm nào phát hành ứng dụng nhiều nhất
df_google_playstore['year'] = df_google_playstore['Released'].apply (lambda x: str(x).split('-')).str[0]
df_google_playstore.groupby(['year']).count()['Released'].sort_values(ascending =False)
```

```python
# Yêu cầu: Tìm hiểu ứng dụng nào lâu rồi chưa cập nhật
df_google_playstore['Last Updated'].sort_values(ascending =True)[:1]
```
